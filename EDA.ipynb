# Exploratory Data Analysis

```python
gender_plot = sns.countplot(x = "Gender", data = df)
for bar in gender_plot.containers:
    gender_plot.bar_label(bar)
```

```python
sales = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending = False)
sales
```

```python
sales_plot = sns.barplot(x = "Gender", y = "Amount", data = sales)
for bar in sales_plot.containers:
    sales_plot.bar_label(bar)
```

### Sales coming from each age group

```python
ax = sns.countplot(x = "Age Group", hue="Gender", data = df)
```

```python
ax = sns.countplot(x = "Age Group", data = df)
```

### Top 10 states with highest number of orders

```python
state_order = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders", ascending = False).head(10)
```

```python
state_order
```

```python
sns.barplot(data = state_order, x = "State", y = "Orders")
sns.set(rc={"figure.figsize":(18,8)})
```

### Top 10 states with highest purchase amount

```python
state_amount = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)
```

```python
state_amount
```

```python
sns.barplot(data = state_amount, x = "State", y = "Amount")
```

```python
ms = df.groupby(["Marital_Status", "Gender"],as_index = False)["Orders"].sum().sort_values(by="Orders").head(10)
```

```python
ms
```

```python
sns.barplot(data = ms, x = "Gender", y = "Orders", hue = "Marital_Status")
sns.set(rc={"figure.figsize":(12,5)})
```

```python
ms_a = df.groupby(["Marital_Status", "Gender"],as_index = False)["Amount"].sum().sort_values(by="Amount").head(5)
```

```python
sns.barplot(data = ms_a, x = "Gender", y = "Amount", hue = "Marital_Status")
```

### Occupation

### Product Category

```python
product_order = df.groupby(["Product_Category"],as_index=False)["Orders"].sum().sort_values(by="Orders", ascending = False).head(10)
sns.set(rc={"figure.figsize":(20,7)})
sns.barplot(product_order, x = "Product_Category", y = "Orders")
```

```python
product_order = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)
sns.set(rc={"figure.figsize":(20,7)})
sns.barplot(product_order, x = "Product_Category", y = "Amount")
```
